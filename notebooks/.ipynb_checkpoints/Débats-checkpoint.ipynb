```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import datetime
from folium.plugins import HeatMap
```

```python
events = pd.read_csv('EVENTS.csv', index_col=0)
```

```python
dates = pd.to_datetime(pd.to_datetime(events.startAt).apply(lambda x: str(x.date())))
```

```python
plt.figure(figsize=(16, 6))
dates.groupby(dates.values).count().sort_index()['2019-01':'2019-03'].plot()
```

```python
weekdays = dates.apply(lambda x: x.weekday())
weekdays = weekdays.groupby(weekdays.values).count()
weekdays = pd.Series(weekdays.values, index=['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche'])
```

```python
plt.bar(weekdays.index, weekdays.values)
```
