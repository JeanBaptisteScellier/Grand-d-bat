```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
import datetime
from folium.plugins import HeatMap
```

```python
events = pd.read_csv('../data/EVENTS.csv', index_col=0)
```

```python
# Trouve la dernière occurence d'un nombre à 4 ou 5 chiffres dans l'addresse
zipcode = events.fullAddress.apply(lambda x: str(x)[::-1]).str.extract(r'(\d{5})')[0].dropna().apply(lambda x: x[::-1])
zipcode = zipcode.append(events.fullAddress.str.extract(r'(\ \d{4}\ )')[0].dropna().apply(lambda x: '0'+x[1:-1]))
# Pour avoir les différents DOM
transco = {'971':'9A', '972':'9B', '973':'9C', '974':'9D', '976':'9F', '200':'2A', '201':'2A', '202':'2B', '206':'2B'}
dep = zipcode.apply(lambda x: x[:3]).apply(lambda x: x[:2] if x not in transco.keys() else transco[x])
# On enlève les codes postaux non français
dep = dep.drop(dep.loc[(dep=='98')|(dep=='99')].index)
departements = dep.groupby(dep.values).count()
```

```python
dates = pd.to_datetime(pd.to_datetime(events.startAt).apply(lambda x: str(x.date())))
```

```python
plt.figure(figsize=(16, 6))
dates.groupby(dates.values).count().sort_index()['2019-01':'2019-03'].plot()
plt.show()
```

```python
weekdays = dates.apply(lambda x: x.weekday())
weekdays = weekdays.groupby(weekdays.values).count()
weekdays = pd.Series(weekdays.values, index=['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche'])
```

```python
plt.figure(figsize=(16, 5))
plt.bar(weekdays.index, weekdays.values)
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(departements.index[:50], departements.values[:50])
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(departements.index[50:], departements.values[50:])
plt.show()
```

```python
heat_df = events.loc[:, ['lat', 'lng']]
heat_df = heat_df.dropna(axis=0, subset=['lat','lng'])
heat_data = [[row['lat'],row['lng']] for index, row in heat_df.iterrows()]
```

```python
width, height = 980, 700
fig = folium.element.Figure(width=width, height=height)
hmap = folium.Map(location=[46.5, 2], zoom_start=6)

hm_wide = HeatMap(heat_data, radius=13, max_zoom=10).add_to(hmap)

fig.add_children(hmap)
```

```python
pop = pd.read_csv('../data/Population_par_departement.csv', encoding="ISO-8859-1", sep=';', names=['Département', 'Population', '', '']).iloc[:-2, :2]
pop['Code'] = pop['Département'].apply(lambda x: x[1:3])
pop['Département'] = pop['Département'].apply(lambda x: x[4:])
pop = pop.set_index('Code')
pop = pop.astype({'Département':'object', 'Population':'int64'})
```

```python
# Nombre de réunions pour 100,000 habitants
ratio = departements/pop['Population']*100000
```

```python
plt.figure(figsize=(16, 10))
plt.bar(ratio.index[:50], ratio.values[:50])
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.bar(ratio.index[50:], ratio.values[50:])
plt.show()
```

```python
pop['Ratio'] = ratio
```

```python
pop.set_index('Département')['Ratio'].sort_values(ascending=False)
```
